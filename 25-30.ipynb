```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/dsuji/Downloads/matches.csv")
data
```

```python
print("Columns:", list(data.columns))
print("\nSample data:\n", data.head())
```

```python
#What percentage of matches were won by the team that chose to field?

field_matches = data[data['Toss_Decision'] == 'field']  
field_wins = field_matches[field_matches['Toss_Winner'] == field_matches['Winners']]
percentage_field_wins = (len(field_wins) / len(field_matches)) * 100
print("Percentage of matches won by team that chose to field:", round(percentage_field_wins, 2), "%")
```

```python
#Which city has the most balanced win distribution among teams?

city_wins = data.groupby(['City','Winners']).size().unstack(fill_value=0)
city_balance = city_wins.max(axis=1) - city_wins.min(axis=1)
print("Most balanced city:", city_balance.idxmin())
```

```python
#Are there any cities where one team dominated completely?

dominant_cities = city_wins[city_wins.apply(lambda x: (x > 0).sum() == 1, axis=1)]
print("\nCities where one team dominated completely:\n", dominant_cities.index.tolist())
```

```python
#What are the overall toss-winning trends for top-performing teams?

top_teams = data['Winners'].value_counts().head(5).index
toss_trends = data[data['Toss_Winner'].isin(top_teams)].groupby('Toss_Winner')['Toss_Decision'].value_counts(normalize=True).unstack(fill_value=0) * 100
print("\nToss-winning trends for top-performing teams (%):\n", toss_trends)
```

```python
#Based on current data, what would be the ideal toss decision for a team playing in each city?

city_toss = data.groupby(['City', 'Toss_Decision'])['Winners'].count().unstack(fill_value=0)
city_toss['ideal_decision'] = city_toss.idxmax(axis=1)
print("\nIdeal toss decision for each city:\n", city_toss[['ideal_decision']])
```
